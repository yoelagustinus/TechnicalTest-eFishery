## Import Packages

```python
import pandas as pd
import numpy as np # linear algebra
import matplotlib.pylab as plt

import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
#import matplotlib.pyplot as plt
```

### Define all Dataset

```python
df_051 = pd.read_excel('dataset/051.xlsx')
df_101 = pd.read_excel('dataset/101.xlsx')
df_102 = pd.read_excel('dataset/102.xlsx')
df_103 = pd.read_excel('dataset/103.xlsx')
df_161 = pd.read_excel('dataset/161.xlsx')
df_162 = pd.read_excel('dataset/162.xlsx')
df_163 = pd.read_excel('dataset/163.xlsx')
```

### Mengubah pseudo-timestamp menjadi data time yang bisa di baca
asumsi: timestamp yang digunakan dengan satuan unit ms (milisecond)

```python
df_051['timestamp'] = pd.to_datetime(df_051['timestamp'], unit='ms')
df_101['timestamp'] = pd.to_datetime(df_101['timestamp'], unit='ms')
df_102['timestamp'] = pd.to_datetime(df_102['timestamp'], unit='ms')
df_103['timestamp'] = pd.to_datetime(df_103['timestamp'], unit='ms')
df_161['timestamp'] = pd.to_datetime(df_161['timestamp'], unit='ms')
df_162['timestamp'] = pd.to_datetime(df_162['timestamp'], unit='ms')
df_163['timestamp'] = pd.to_datetime(df_163['timestamp'], unit='ms')
```

## Merge all Datasets be 1 Dataset

```python
merged_df = pd.concat([df_051, 
df_101,
df_102,
df_103,
df_161,
df_162,
df_163])

# , ignore_index=True
```

```python
merged_df
```

### Save new dataframe to new csv

```python
merged_df.to_csv("dataset/merge_data.csv", index=False)
```

```python
df = pd.read_csv('dataset/merge_data.csv')
```

### Mencari data duplicate dengan Axis x, y, z

```python
# merged_df.duplicated(subset=['x', 'y', 'z'])

duplicates = merged_df[merged_df.duplicated(subset=['x', 'y', 'z'], keep=False)].groupby(['x','y','z'])
```

### Print data sesuai Axis yang sama 

```python
# menampilkan data yang memiliki nilai yang sama pada kolom 'x,y,z'
for axis, group in duplicates:
    print(f'Axis: {axis}:')
    group.sort_values('timestamp', ascending=True)
    print(group)
```

## 2. Jika anda telah mengetahui durasi makan ikan, dan pakan diberikan secara otomatis menggunakan autofeeder, jadwal yang digunakan untuk melakukan feeding seperti apa? (Asumsi: pakan yang diberikan seberat 5 kg pada setiap event makan)
Answer: <br>

Jika pakan yang diberikan seberat 5Kg pada setiap event makan. Menurut saya setiap jumlah ikan di setiap lokasi Axis (x,y,z) berjumlah 15% dari keseluruhan ikan. Maka, lokasi Axis(x,y,z) tersebut hanya mendapatkan jumlah 15% dari 5Kg Pakan yang berarti bernilai <b>750g</b> ((15/100)*5000g)
