```python



# Convert time column to datetime
#data['time'] = pd.to_datetime(data['time'])

# Set time column as index
df_051 = df_051.set_index('timestamp')

# Plot the data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_051['x'], df_051['y'], df_051['z'])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python

plt.figure()
plt.subplots(figsize=(16, 8))
plt.plot(df_051['x'], color = 'green', label = 'x')
plt.plot(df_051['y'], color = 'blue', label = 'y')
plt.plot(df_051['z'], color = 'red', label = 'z')
#plt.title("")
plt.xlabel('Time')
plt.ylabel("Axis")
plt.legend()

plt.show()
```
