## Merge all Datasets be 1 Dataset

```python
merged_df = pd.concat([df_051, 
df_101,
df_102,
df_103,
df_161,
df_162,
df_163], ignore_index=True)

```

```python
# asc by timestamp
merged_df.sort_values('timestamp',ascending=True)
```

```python
# merged_df.duplicated(subset=['x', 'y', 'z'])

duplicates = merged_df[merged_df.duplicated(subset=['x', 'y', 'z'], keep=False)].groupby(['x','y','z'])
```

```python
# menampilkan data yang memiliki nilai yang sama pada kolom 'x,y,z'
for axis, group in duplicates:
    print(f'Axis: {axis}:')
    print(group)
```
